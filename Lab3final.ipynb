```python
import pandas as pd
x='diabetes_dataset.csv'
df = pd.read_csv(x)
```

```python
print(df)
```

```python
import numpy as np
arr=df.values
X=arr[:,0:8]
Y=arr[:,8]
```

```python
#Partitioning and K-Fold Cross Validation

from sklearn.model_selection import KFold
from sklearn.model_selection import RepeatedKFold 
kf=RepeatedKFold(n_splits=5,n_repeats = 10,random_state = None)
for train_index,test_index in kf.split(X):
  print("Train:",train_index,"\nValidation:",test_index)
  X_train,X_test = X[train_index],X[test_index]
  Y_train,Y_test = Y[train_index],Y[test_index]
```

```python
print(len(X_train))
print(len(X_test))
```

```python
#Stratified KFold

from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits = 5,random_state = None)
for train_index,test_index in skf.split(X,Y):
  print("Train:",train_index,"\nTest",test_index)
  X_train,X_test = X[train_index],X[test_index]
  Y_train,Y_test = Y[train_index],Y[test_index]
```

```python
#Simple KFold with Repeatition

rkf = RepeatedKFold(n_splits = 5,n_repeats=10,random_state=None)
for train_index,test_index in rkf.split(X,Y):
  print("Train:",train_index,"\nTest:",test_index)
  X_train,X_test = X[train_index],X[test_index]
  Y_train,Y_test = Y[train_index],Y[test_index]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
from sklearn.metrics import auc
from sklearn.metrics import roc_auc_score
max = -1
k=[1,3,5,7,9,11,13,15]
for i in k:
  classifier = KNeighborsClassifier(n_neighbors = i)
  classifier.fit(X_train,Y_train)
  Y_pred = classifier.predict_proba(X_test)
  fpr,tpr,threshold = roc_curve(Y_test,Y_pred[:,1])

  print(fpr)
  print(tpr)
  print(threshold)

  roc_auc = auc(fpr,tpr)
  print("ROC_AUC Score: "+str(roc_auc))

  roc_auc_dir = roc_auc_score(Y_test,Y_pred[:,1])
  print("ROC_AUC_Direct Score: " +str(roc_auc_dir))

  if(max<roc_auc):
    max = roc_auc
    maxK = i

  plt.style.use('seaborn')

  plt.legend(loc='lower right')
  plt.plot([0,1],[0,1],'r--',label = 'Base Model')
  plt.xlim([0,1])
  plt.ylim([0,1])
  plt.ylabel('True Positive Rate')
  plt.xlabel('False Positive Rate')
  plt.title('K = '+str(i))

  plt.plot(fpr,tpr,linestyle='--',label='AUC = %0.2f' %roc_auc)
  plt.legend(loc = 'lower right')
  plt.show()
```

```python
print('The KNN model that performs better is '+str(maxK))
print('The ROC_AUC score of KNN Model is '+str(max))
```
